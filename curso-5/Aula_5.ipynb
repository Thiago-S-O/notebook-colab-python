```python
# importando da biblioteca as funções create_engine, MetaData, Table, inspect
'''
A primeira importação será create_engine, que serve para criar o motor, ou seja, a conexão com 
o banco de dados que o fará funcionar.

Na mesma linha, inserimos uma vírgula e a importação MetaData. Essa é uma classe que representa 
os meta dados do banco de dados, isso inclui informações sobre as tabelas, colunas e outros objetos.

Também importamos a classe Table que representa uma tabela em um banco de dados e é usada para 
manipular dados em uma tabela específica.

Por fim, importamos a função inspect, que permite inspecionar a estrutura do banco de dados.
'''
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Nesse caso, usaremos o SQLite, um banco de dados que vem naturalmente instalado no Google Colab. 
# Nesse trecho de código o memory indica que estamos usando um banco de dados local.
# salvo na variável 'engine'
engine = create_engine('sqlite:///:memory:')
```

###**Escrevendo em um banco de dados**

```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# salvando a url dos dados lá da github em um variável  
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
# lendo o arquivo csv da url e salvando em uma variável
dados = pd.read_csv(url)
```

```python
# lendo as primeiras linhas
dados.head()
```

```python
# convertendo esses dados para o formato de sql, passando os seguintes parâmetros: nome_da_tabela, motor do banco, 
# e retirar os indices que vem junto com os dados do formato csv
dados.to_sql('clientes', engine, index=False)
```

```python
# chamando a função inspect() para verificar o que tem no banco de dados e inspecionar algumas informações;
# e passando o motor do banco de dados dentro da função;
# salvando dentro de uma variável.
inspector = inspect(engine)
```

```python
# para printar o nome das tabelas que tem dentro do banco de dados. é necessário passar o inspector.get_table_names() dentro 
# da função print(), onde essa função devolve os nomes das tabelas que estão dentro do banco de dados.
# é possível observar que o nome da tabela é o mesmo que foi salvo acima, logo está funcionando corretamente.
print(inspector.get_table_names())
```

###**Lendo uma consulta SQL**

```python
# fazendo uma consulta nos dados da tabela 'cliente', que nem como se fosse fazer uma consulta no banco de dados do SQLSERVER
# onde é filtrado na coluna 'Categoria_de_renda' o 'Empregrado'
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
# salvando a consulta em uma vairável chamada 'empregrados', utilizando o método read_sql() da blbioteca pandas
# onde é passado os parâmetros da consulta, no caso, a variável query cona consulta, e o motor do banco de dados
empregados = pd.read_sql(query, engine)
```

```python
# printando a consulta
empregados
```

```python
# salvando so dados da consulta em uma tabela no formato sql
# o 'con' é utlizando para especificar o objeto engine do sqlalchemy que foi criando 
# anteriormente para se conectar ao banco de dados
empregados.to_sql('empregados', con=engine, index=False)
```

```python
# para ler a tabela inteira salva acima, é utilizada a função pd.read_sql_table()
# onde é passados os o nome da tabela a ser executada, no caso a 'empregados', e o engine
#
# o read_sql permite ler uma consulta e o read_sql_table uma tabela completa.
pd.read_sql_table('empregados', engine)
```

```python
# fazendo um filtro dos dados a serem lidos das tabela salva, 
# adicionando o parâmetro 'columns' e passando o nome das colunas a serem lidas
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade',
                                                 'Rendimento_anual'])
```

###**Atualizando um banco de dados**

```python
# fazendo uma consulta na tabela 'clientes', assim como uma consuta no banco de dados no sql
query = 'SELECT * FROM clientes'
```

```python
# lendo a consulta
pd.read_sql(query, engine)
```

```python
# no caso do vscode, precisa converter a quary de uma simples string para texto, pois o SQLAlchemy 
# espera um objeto SQL executável (como uma TextClause ou um objeto compilado de text()) em vez de 
# uma simples string ao executar consultas diretamente.
from sqlalchemy import text

# deletando uma informação específica por meio de uma query, assim como no sql
query = text('DELETE FROM clientes WHERE ID_Cliente=5008804')

# usando um método chamado 'connect' para estabelecer uma conexão com o banco de dados, é feito um pedindo 
# para executar a query com o método execute que quando chamado, executa a instrução SQL diretamente no 
# banco de dados, usando a conexão estabelecida.
with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
# text(): Converte sua string SQL em um objeto executável.
# commit(): Em operações DELETE, INSERT ou UPDATE, você precisa confirmar explicitamente as mudanças 
# feitas no banco de dados.
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(query)
```

```python
# lendo a consulta atualizada
pd.read_sql_table('clientes', engine)
```

```python
# alterando uma informação por meio de uma consulta (execução no Google Colab)
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(query)
```

```python
# alterando uma informação por meio de uma consulta (execução no vscode)
from sqlalchemy import text

# a consulta é da mesma forma como no SQLSERVER
query = text('UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808')

with engine.connect() as conn:
    conn.execute(query)
    conn.commit()
```

```python
# verificando a nova informação atualizada
pd.read_sql_table('clientes', engine)
```
